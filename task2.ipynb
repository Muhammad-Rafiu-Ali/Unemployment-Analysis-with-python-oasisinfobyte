# Dataset: Unemployment Analysis with python

```python
# Step 1: Load necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Step 2: Load the dataset
dataset_path = r"C:\Users\PMYLS\Desktop\ossis intern\archive (4)\Unemployment_Rate_upto_11_2020.csv"
unemployment_data = pd.read_csv(dataset_path)
```

```python
print(unemployment_data.columns)
```

```python
# Step 4: Exploratory Data Analysis (EDA)
# Visualize the unemployment rate over time
plt.figure(figsize=(10, 6))
sns.lineplot(x=' Date', y=' Estimated Unemployment Rate (%)', data=unemployment_data)
plt.title('Unemployment Rate Over Time')
plt.xlabel('Date')
plt.ylabel('Estimated Unemployment Rate (%)')
plt.grid(True)
plt.xticks(rotation=45)
plt.show()
```

```python
# Step 5: Statistical Analysis
mean_unemployment_rate = unemployment_data[' Estimated Unemployment Rate (%)'].mean()
median_unemployment_rate = unemployment_data[' Estimated Unemployment Rate (%)'].median()
std_unemployment_rate = unemployment_data[' Estimated Unemployment Rate (%)'].std()

print("Mean Unemployment Rate:", mean_unemployment_rate)
print("Median Unemployment Rate:", median_unemployment_rate)
print("Standard Deviation of Unemployment Rate:", std_unemployment_rate)
```

```python
# Calculate the monthly unemployment rate change
unemployment_data['Unemployment Rate Change'] = unemployment_data[' Estimated Unemployment Rate (%)'].diff()

# Print the first few rows to verify the calculation
print(unemployment_data[[' Date', ' Estimated Unemployment Rate (%)', 'Unemployment Rate Change']].head())
```

```python
# Visualize the monthly change in unemployment rate
plt.figure(figsize=(10, 6))
sns.lineplot(x=' Date', y='Unemployment Rate Change', data=unemployment_data)
plt.title('Monthly Change in Unemployment Rate')
plt.xlabel('Date')
plt.ylabel('Change in Unemployment Rate (%)')
plt.grid(True)
plt.xticks(rotation=45)
plt.axhline(y=0, color='r', linestyle='--')  # Add a horizontal line at y=0
plt.show()
```

```python
# Step 7: Comparison and Benchmarking
# Compare unemployment rates across different regions if available
plt.figure(figsize=(10, 6))
sns.boxplot(x='Region', y=' Estimated Unemployment Rate (%)', data=unemployment_data)
plt.title('Comparison of Unemployment Rates Across Regions')
plt.xlabel('Region')
plt.ylabel('Estimated Unemployment Rate (%)')
plt.xticks(rotation=45)
plt.show()
```
